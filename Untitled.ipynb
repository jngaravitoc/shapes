```python
%pylab inline
font = {'size':20, 'family':'serif'}
matplotlib.rc('font', **font)
```

```python
data = np.loadtxt('MW1LMC4a1H6shape.txt', skiprows=1)
```

```python
q = data[:,0]
s = data[:,1]
```

```python
figure(figsize=(8, 6))
time = linspace(0, 2, 11)
plot(time, q, lw=3, label=r'$\frac{b}{a}$', ls='-.')
plot(time, s, lw=3, label=r'$\frac{c}{a}$', ls='--')
plot(time, q/s, lw=3, label=r'$\frac{b}{c}$', ls='-')
legend(fontsize=30, ncol=1, loc='best')
xlabel('Time(Gyrs)', fontsize=25)
title('$c\leq b \leq a$', fontsize=35)
savefig('shape_evolution.png', bbox_inches='tight')
```
